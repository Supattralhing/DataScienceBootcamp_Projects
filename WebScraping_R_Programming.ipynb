HW01: Start web scraping IMDB web

```r
install.packages("rvest") 
install.packages("tidyverse")
```

Load library

```r
library(rvest)
library(tidyverse)
```

```r
url <- "https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc"

movie_name <- url %>%
  read_html() %>%
  html_elements("h3.lister-item-header") %>%
  html_text2()
movie_name_new <- gsub("\\s*\\([^\\)]+)", "", movie_name) #remove text within ()

ratings <- url %>%
  read_html() %>%
  html_elements("div.ratings-imdb-rating") %>%
  html_text2() %>%
  as.numeric()

genre <- url %>%
  read_html() %>%
  html_elements("span.genre") %>%
  html_text2()

length <- url %>%
  read_html() %>%
  html_elements("span.runtime") %>%
  html_text2()

year <- url %>%
  read_html() %>%
  html_elements("span.lister-item-year") %>%
  html_text2()
released_year <- gsub("[()]", "", year)#remove ()

df_movies <- data.frame(movie_name_new, released_year, genre, ratings, length)
View(df_movies)
```

HW02: Start web scraping static web

```r
url <- "https://readery.co/best-sellers"

book_name <- url %>%
  read_html() %>%
  html_elements("div.book-title a.title") %>%
  html_text2()

author <- url %>%
  read_html() %>%
  html_elements("div.book-title a.author") %>%
  html_text2()

list_prices <- url %>%
  read_html() %>%
  html_elements("span.listprice del") %>%
  html_text2()

sales_prices <- url %>%
  read_html() %>%
  html_elements("span.saleprice") %>%
  html_text2() 

df_books <- data.frame(book_name, author, list_prices, sales_prices)
View(df_books)
```
